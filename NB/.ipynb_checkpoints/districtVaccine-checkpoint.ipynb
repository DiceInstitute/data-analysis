```python
#Importing the library for Panda
import pandas as pd
```

```python
#Reading csv from panda read csv
df=pd.read_csv('../CovidFirstDoseAnalysis/vaccine_dose_data_district.csv')
```

```python
#Checking the df data
df
```

```python
#Collecting all the columns which we need to analyze
mask=df.loc[0].str.startswith('First Dose Administered').fillna(True)
```

```python
df1=df.T.loc[mask].T.drop(0)
```

```python
#Checking the data
df1
```

```python
#Write it to an excel
df1.to_excel('../CovidFirstDoseAnalysis/FirstDoseDistrict.xlsx')
```

```python
#Deleting the junk character in columns
df1.columns=df1.columns.astype(str).str.split('.').str[0]
```

```python
#Write to an excel
df1.to_excel('../CovidFirstDoseAnalysis/FirstDoseDistrict.xlsx')
```
