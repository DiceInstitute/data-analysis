```python
#Importing for the data scraping from yahoo finance.
import pandas_datareader.data as reader

#Importing for the reading and writing the excel as panda data frame
import pandas as pd

#Importing for the date time
import datetime as dt

#Importing for plotting the points in XY plane
import matplotlib.pyplot as plt

#Importing for warnings
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)

#Importing to use the time series concept
from prophet import Prophet
```

```python
#Scraping historical data for Ethereium and saving it to the variable eth
eth = reader.get_data_yahoo('ETH-USD')

#Saving the data into an excel file ethereium.xlsx
eth.to_excel("ethereium.xlsx")

#Reading the excel file from the saved excel file
ethusd=pd.read_excel('ethereium.xlsx')

#Adding new columns ds and y from the existing columns Date & Adj Close
ethusd[["ds","y"]] = ethusd[["Date","Adj Close"]]

#Initilizing the prophet model
model=Prophet()

#Fitting the model with the data from the Data Frame ethusd
model.fit(ethusd)

#Predicting the price in future based on the historical data for n days.
future = model.make_future_dataframe(periods=180)

#Assigning the future model to forecast variable
forecast = model.predict(future)

#Plotting into the XY plane for forecasting
model.plot(forecast)

#Naming xy plane
plt.xlabel("Time Period")
plt.ylabel("Price In USD")

#Showing the plot
plt.show()
```
